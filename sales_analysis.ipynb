```python
import pandas as pd

df = pd.read_csv("/sales_analysis.csv")
print(df)
```

```python
sales = df.groupby("salesperson")["total_sales"].sum().sort_values(ascending=False)
print(sales)
```

```python
total_revenue = df['total_sales'].min()
print("Total Revenue:", total_revenue)
```

```python
total_revenue = df['total_sales'].sum()
print("Total Revenue:", total_revenue)
```

```python
region_sales = df.groupby('region')['total_sales'].sum().sort_values(ascending=False)
print(region_sales)
```

```python
import matplotlib.pyplot as plt

region_sales.plot(kind='bar', title='Sales by Region')
plt.xlabel("Region")
plt.ylabel("Revenue")
plt.show()
```

```python
top_products = df.groupby('product')['total_sales'].sum().sort_values(ascending=False).head(5)
print(top_products)
```

```python
top_products.plot(kind='bar', title='Top 10 Products by Revenue')
plt.show()
```

```python
category_sales = df.groupby('category')['total_sales'].sum()
print(category_sales)

```

```python
salesperson_perf = df.groupby('salesperson')['total_sales'].sum().sort_values(ascending=False)
print(salesperson_perf.head())
```

```python
monthly_sales = df.groupby('order_date')['total_sales'].sum()

monthly_sales.plot(kind='line', title='Monthly Sales Trend')
plt.xlabel("order_date")
plt.ylabel("Revenue")
plt.show()
```
